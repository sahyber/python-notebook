# 7. Tuplas

Tuplas são coleções ordenadas e imutáveis. Existem 2 métodos para tuplas: count e index. Podemos criar tuplas com ()

```python
tupla = (1, 2, 3, 4, 5)
type(tupla) # <class 'tuple'> - Tuplas são imutáveis, ou seja, seus elementos não podem ser alterados após a criação.
```

```python
tupla = 1 , 2, 3, 4, 5
type(tupla) # <class 'tuple'> - Tuplas são imutáveis, ou seja, seus elementos não podem ser alterados após a criação.
```

```python
tupla = (1)
type(tupla) # <class 'int'> - Para criar uma tupla com um único elemento, é necessário incluir uma vírgula após o elemento.
```

```python
tupla = (1, )
type(tupla) # <class 'tuple'> - Agora é uma tupla com um único elemento.
```

```python
tupla = 1, 
type(tupla) # <class 'tuple'> - A vírgula é o que define a tupla, não os parênteses.
```

## Métodos em tuplas

| Método | Descrição |
| :--- | :--- |
| index() | Retorna o índice do elemento especificado |
| count() | Conta o número de vezes que o elemento aparece na tupla |

```python
tupla = ('a', 'b', 'c', 'd', 'e')

print(f"Índice do elemento 'c': {tupla.index('c')}") # Retorna o índice do primeiro elemento com o valor 'c' na tupla.
print(f"Índice do elemento 'd': {tupla.index('d')}") # Retorna o índice do primeiro elemento com o valor 'd' na tupla.
```

```python
tupla = ('a', 'b', 'b', 'c', 'd', 'e')

print(f"Número de ocorrências do elemento 'b': {tupla.count('b')}") # Retorna o número de ocorrências do elemento 'b' na tupla, que é 2.
```
