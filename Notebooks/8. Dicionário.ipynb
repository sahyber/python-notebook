# 8. Dicionários

* Tipo de dado chave-valor, ou seja, para cada chave associa um valor correspondente

* Usamos dicionários para criar DataFrames na biblioteca Pandas (DataScience)

```python
dic1 = {
    'nome': 'Alice',
    'idade': 30
}
dic1
```

```python
type(dic1) # <class 'dict'> - Dicionários são estruturas de dados que armazenam pares de chave-valor.
```

```python
dic1['nome'] # 'Alice' - Acessa o valor associado à chave 'nome' no dicionário.
```

```python
dic1['nome'] = 'Bob' # Atualiza o valor associado à chave 'nome' para 'Bob'.
dic1 # O dicionário agora tem 'nome' atualizado para 'Bob'.
```

```python
dic2 = {
    'chave1': [1, 2, 3],
    'chave2': {'subchave': 'valor'},
    'chave3': (4, 5, 6)
}
dic2
```

```python
dic2['chave1'] # [1, 2, 3] - Acessa o valor associado à chave 'chave1', que é uma lista.
```

```python
dic2['chave2']['subchave'] # 'valor' - Acessa o valor associado à chave 'subchave' dentro do dicionário associado à chave 'chave2'.
```

```python
dic2['chave3'][2]
```
