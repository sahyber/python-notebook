# 5. Operadores

* Operadores de Comparação (==, !=, <=, >=, >, <, >)

```python
# Igualdade
10 == 10
```

```python
# Desigualdade (diferente)
10 != 10
```

```python
# Maior ou igual, menor ou igual

print(10 >= 10)
print(10 <= 10)
```

```python
# Maior que

print(10 > 10)
print(10 < 10)
```

```python
string1 = "Python"
string2 = "Java"

# Comparação de comprimento das strings
resultado1 = len(string1) > len(string2)
resultado2 = len(string1) < len(string2)

print(len(string1) == len(string2))
print(resultado1)
print(resultado2)
```

* Operadores Lógicos

| Tabela AND | |
| :--- | :--- |
| VV | V |
| VF | F |
| FV | F |
| FF | F |

```python
# Tabela AND
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

| Tabela OR | |
| :--- | :--- |
| VV | V |
| VF | V |
| FV | V |
| FF | F |

```python
# Tabela OR
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

```python
x = 10
```

```python
print((x > 5) and (x < 15)) # O operador lógico and retorna True se ambas as condições forem verdadeiras
print((x > 15) or (x > 5)) # O operador lógico or retorna True se pelo menos uma das condições for verdadeira
```
