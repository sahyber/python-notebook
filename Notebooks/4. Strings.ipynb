# 4. Strings

```python
# O operador len() é usado para obter o comprimento (número de caracteres) de uma string.
nome = "Samanta"
print(len(nome))

print(len("User"))
```

```python
# O operador in é usado para verificar se uma substring está presente em uma string.

a = "Python is a language"

print("Python" in a)
print("Java" in a)

print("language" not in a)
print("Samanta" not in a)
```

```python
nome = "Carlos"

print(nome.lower()) # O método lower() retorna a string em letras minúsculas
print(nome.upper()) # O método upper() retorna a string em letras maiúsculas
```

```python
# O método strip() remove os espaços em branco do início e do fim da string

espaco = "   Olá, Mundo!   "
print(espaco)

print(espaco.strip())
```

```python
# O método title() retorna a string com a primeira letra de cada palavra em maiúscula

nome = "joão silveira pinheiro"
print(nome.title())
```

```python
nome = "     joão silveira pinheiro     "
print(nome.title().strip())
```

```python
nome.index("a") # O método index() retorna o índice da primeira ocorrência da substring especificada
```

```python
# Cada elementos da string tem um índice (ocupa uma posição), começando do 0
nome = "gabriel"

print(nome[0])
print(nome[2])
```

```python
# Fatiamento de string (slicing)
# O fatiamento é uma técnica que permite extrair partes de uma string usando índices.

print(nome[:2])
print(nome[0:5])

print(nome[5:])
```
