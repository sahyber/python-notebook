# 3. Números

## Tipos numéricos

* int
* float
* complex

```python
x = 10 # int
y = 10.5 #float
z = 10+1j #complex

print(type(x))
print(type(y))
print(type(z))
```

* Conversão (casting)

```python
idade = int(input("Digite sua idade: "))
peso = float(input("Digite seu peso em kg: "))

print("Sua idade é {} anos e seu peso é {} kg.".format(idade, peso))

print(type(idade))
print(type(peso))
```

```python
numero = "10"
print(type(int(numero)))
```

```python
numero = int(15)
print(type(numero))

numero = float(numero)
print(type(numero))
```

* Usando o Python como calculadora

```python
adicao = 10 + 3
subtracao = 10 - 3
multiplicacao = 10 * 3
divisao = 10 / 3
divisao_inteira = 10 // 3
modulo = 10 % 3 # O operador de módulo retorna o resto da divisão inteira
exponenciacao = 10 ** 3

print("Adição:", adicao)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
print("Divisão Inteira:", divisao_inteira)
print("Módulo:", modulo)
print("Exponenciação:", exponenciacao)
```
