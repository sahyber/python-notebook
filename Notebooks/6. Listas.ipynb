## Fatiamento de listas (slicing)

Considere uma lista hipotética z. Se quisermos selecionar todos os valores entre os índices m-n (selecionar todos os valores começando no valor m até o valor de índice n-1):

* z[m:n]

```python
z = [3,75,2,87,45,12,67,34,89,23]
z[1:4]
```

```python
z[1:]
```

```python
z[:5]
```

## Métodos em listas

| Métodos | Descrição |
| :--- | :--- |
| append() | Adiciona um elemento ao final da lista |
| insert() | Adiciona um elemento em um índice especificado |
| extend() | Adiciona os elementos de um iterável(lista, tupla, ...) no final da lista |
| pop() | Remove e retorna um elemento de um índice especificado |
| remove() | Remove o primeiro elemento com o valor especificado |
| clear() | Remove todos os elementos de uma lista |
| sort() | Ordena uma lista |
| copy() | Retorna uma cópia da lista selecionada |
| reverse() | Reverte a ordem dos elementos em uma lista |
| index() | Retorna o primeiro índice do elemento especificado |
| count() | Conta o número de vezes que o elemento aparece na lista |

append() -> Adiciona um elemento ao final da lista

```python
numeros = [45, 67, 89, 12, 34]
numeros.append(56)
print(numeros)

# Recebe apenas um elemento por vez, que é adicionado ao final da lista.
numeros.append(78)
print(numeros)
```

```python
# Existe a alternativa de colocar os números dentro de uma lista ou tupla e passar como arqumento da função append
# mas isso não é recomendado, pois o resultado será uma lista aninhada, ou seja, uma lista dentro de outra lista.

numeros.append([90, 100])
print(numeros)
```

extend() -> Adiciona os elementos de um iterável(lista, tupla, ...) no final da lista

```python
numeros = [0, 0, 0, 0, 0]
numeros.extend([1, 2, 3, 4, 5]) # Adicionamos apenas os números do iterável (lista) sem criar uma lista aninhada.
print(numeros)
```

```python
numeros.extend('abcde') # Adiciona cada caractere da string como um elemento separado na lista.
print(numeros)
```

insert() -> Adiciona um elemento em um índice especificado

Síntaxe:

* lista.insert(index.object) - primeiro passamos o índice e depois o objeto

```python
numeros = [4, 6, 8, 10, 12]
numeros[1]
```

```python
numeros.insert(1, 5) # Insira o número 5 na posição de índice 1 (entre 4 e 6).
print(numeros)
```

remove() -> Remove o primeiro elemento com o valor especificado

ValueError: se inserirmos um valor que não esteja na lista teremos esse tipo de erro

```python
numeros = [32, 45, 67, 89, 12]
numeros.remove(45) # Remove o número 45 da lista.
print(numeros)
```

pop() -> Remove e retorna um elemento de um índice especificado

```python
lista = ['java', 'python', 'c++', 'javascript']
lista.pop() # Remove o último elemento da lista e o retorna.
print(lista)
```

```python
lista.pop(1) # Remove o elemento no índice 1 (python) e o retorna.
print(lista)
```

Também podemos deletar elementos de uma lista com a instrução del:

```python
letras = ['a', 'b', 'c', 'd', 'e']
del letras[2] # Remove o elemento no índice 2 (c) da lista.
print(letras)
```

Podemos deletar completamente a lista:

```python
del letras # Remove a lista inteira da memória.

# Já que a lista foi deletada, tentar acessá-la resultará em um erro.
```

clear() -> Remove todos os elementos de uma lista

```python
letras = ['a', 'b', 'c', 'd', 'e']
letras
```

```python
letras.clear() # Remove todos os elementos da lista, mas mantém a lista em si.
print(letras)
```

count() -> Conta o número de vezes que o elemento aparece na lista

```python
numeros = [32, 45, 67, 89, 12, 45, 89, 45]
numeros.count(45) # Conta quantas vezes o número 45 aparece na lista.
```

```python
numeros.count(89)
```

reverse() -> Reverte a ordem dos elementos em uma lista

```python
numeros = [32, 45, 67, 89, 12]
numeros.reverse() # Inverte a ordem dos elementos na lista.
print(numeros)
```

sort() -> Ordena uma lista

```python
numeros = [32, 45, 67, 89, 12]
numeros
```

```python
numeros.sort() # Ordena os elementos da lista em ordem crescente.
numeros
```

index() -> Retorna o primeiro índice do elemento especificado

* index(value, start=0, stop)

```python
lista = [1, 2, 3, 4, 5]
```

```python
lista.index(4) # Retorna o índice do primeiro elemento com o valor 4 na lista.
```

```python
lista = [1, 2, 3, 3, 3, 3, 3, 4, 5]
lista.index(3) # Retorna o índice do primeiro elemento com o valor 3 na lista.
```

```python
lista.index(3, 4) # Retorna o índice do primeiro elemento com o valor 3 na lista, começando a busca a partir do índice 4.
```
